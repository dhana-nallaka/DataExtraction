```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL='https://www.amazon.com/s?k=Apple+Watches&crid=36UJ5NW1BH78N&sprefix=apple+watche%2Caps%2C163&ref=nb_sb_noss_2'
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
```

```python
webpage = requests.get(URL, headers=HEADERS)
webpage
```

```python
type(webpage.content)
```

```python
soup=BeautifulSoup(webpage.content, "html.parser")
type(soup)
```

```python
links = soup.find_all("a", attrs={'class':'a-link-normal s-line-clamp-2 s-link-style a-text-normal'})
#links
```

```python
link=links[0].get('href')
```

```python
productlist='https://amazon.com'+link
productlist
```

```python
newwebpage = requests.get(productlist, headers=HEADERS)
newwebpage
```

```python
type(newwebpage.content)
```

```python
newsoup=BeautifulSoup(newwebpage.content,"html.parser")
#newsoup
```

```python
collecting_data=newsoup.find("span",attrs={"id":'productTitle'})
```

```python
finaldata=collecting_data.text.strip()
finaldata
```

```python
price_partone=newsoup.find("span",attrs={"class":'a-price-whole'}).text
```

```python
price_parttwo=newsoup.find("span",attrs={"class":'a-price-fraction'}).text
```

```python
total_price=(price_partone+price_parttwo)
total_price
```
